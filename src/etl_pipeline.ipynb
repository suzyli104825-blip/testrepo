### transform data

```python
from pyspark.sql.functions import col, current_date, months_between, floor, to_date
#add a new column:withColumn(),  get the integer: floor(),
df_drivers = df_drivers.withColumn(
    "age",
    floor(months_between(current_date(), to_date(col("dob"))) / 12)
)
```

```python
display(df_drivers)
```

```python
df_lap_drivers=df_drivers.select("driverId","driverRef","number","code","forename","surname","dob","nationality","age").join(df_laptimes,on="driverId")
```

```python
display(df_lap_drivers)
```

```python
df_lap_drivers = df_lap_drivers.groupby("age")
```

```python

df_lap_drivers.write.csv("/Volumes/gr5069/ne2388/inclass/sl5869_lap_times_output.csv", header=True)
```
